```python
#scatter plot to visualize relationship between Hours spent studying and Scores
x1, y1 = np.loadtxt('student_scores2.txt', usecols = [0,2], unpack=True, skiprows=1)
plt.plot(x1, y1,'o')
plt.xlabel("Hours Spent Studying")
plt.ylabel("Score")
plt.show()
```

Finding best line of fit to the scatter plot

```python
def line(x1, a, b):
    return a * x1 + b
```

```python
constants = curve_fit(line, x1, y1)
```

```python
a_fit = constants [0][0]
b_fit = constants [0][1]
```

```python
fit = []
for i in x1:
    fit.append(line(i,a_fit,b_fit))
```

```python
plt.plot(x1,y1, 'o')
plt.plot(x1,fit)
plt.xlabel("Hours Spent Studying")
plt.ylabel("Score")
plt.title('Python curve fit')
plt.show()
```

Based on the graph, Hours spent studying and Scores obtained are positiveley correlated

```python
#scatter plot to visualize relationship between IQ and Scores
x2, y2 = np.loadtxt('student_scores2.txt', usecols = [1,2], unpack=True, skiprows=1)
plt.plot(x2, y2,'o')
plt.xlabel("IQ")
plt.ylabel("Score")
plt.show()
```

```python
# This will fit the best line into the graph
plt.scatter(x2,y2)
plt.plot(np.unique(x2), np.poly1d(np.polyfit(x2, y2, 1))
         (np.unique(x2)), color='red')
plt.xlabel("IQ")
plt.ylabel("Score")
```

The graph also shows that the relationship between IQ and Student score are positively correlated

**CORRELATION MATRIX**<br>
This a measure of the linear association between two variables. It has a value between -1 and 1 where:

1. -1 indicates a perfectly negative linear correlation between two variables
2. 0 indicates no linear correlation between two variables
3. 1 indicates a perfectly positive linear correlation between two variables
The further away the correlation coefficient is from zero, the stronger the relationship between the two variables.

```python
corr = data.corr()
corr
```

```python
fig, ax = plt.subplots(figsize=(8, 6))
im = ax.imshow(data.corr(), cmap='coolwarm')

plt.colorbar(im)

ax.set_xticks(range(len(data.columns)))
ax.set_yticks(range(len(data.columns)))
ax.set_xticklabels(data.columns, rotation='vertical')
ax.xaxis.tick_top()
ax.set_yticklabels(data.columns)
plt.title('Correlation Matrix')

plt.show()
```
