```python
 x1, y1 = np.loadtxt('student_scores2.txt', usecols = [0,2], unpack=True, skiprows=1)
```

```python
plt.plot(x1, y1,'o')
plt.xlabel("Hours Spent Studying")
plt.ylabel("Score")
plt.show()
```

```python
def line(x1, a, b):
    return a * x1 + b
```

```python
constants = curve_fit(line, x1, y1)
```

```python
a_fit = constants [0][0]
b_fit = constants [0][1]
```

```python
fit = []
for i in x1:
    fit.append(line(i,a_fit,b_fit))
```

```python
plt.plot(x1,y1, 'o')
plt.plot(x1,fit)
plt.xlabel("Hours Spent Studying")
plt.ylabel("Score")
plt.title('Python curve fit')
plt.show()
```
